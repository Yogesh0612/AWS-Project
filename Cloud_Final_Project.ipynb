```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\user\Downloads\78b5d6d2-d3e3-45d9-84b3-dc90d8cfff82_full_prediction.csv')
```

```python
sns.displot(data['probability'], kind = 'kde')
plt.show()
```

```python
data.drop(['bankrupt', 'company_id', 'company'], axis = 1).corr()
```

```python
corr_matrix = data[['probability','current_liabilities_to_equity',
'current_liability_to_current_assets',
'working_capital_to_equity',
'working_capital_to_total_assets']].corr()
```

```python
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust the font size for better visibility
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    xticklabels=["Probability", "Liabilities to Equity", "Liability to Current Assets", "Working Capital to Equity", "Working Capital to Total Assets"],
    yticklabels=["Probability", "Liabilities to Equity", "Liability to Current Assets", "Working Capital to Equity", "Working Capital to Total Assets"],
)

# Set the title
plt.title("Correlation Heatmap")

# Adjust plot settings
plt.tight_layout()

# Save the heatmap as an image file (e.g., for use in a presentation)
plt.savefig("correlation_heatmap.png", dpi=300)

# Show the plot (optional)
plt.show()
```

```python
corr_columns = ['debt_ratio_percentage', 'equity_to_longterm_liability', 'liability_to_equity']
corr_matrix_debt = data[corr_columns].corr()

# Create a heatmap for the subset
plt.figure(figsize=(8, 6))
sns.set(font_scale=1.2)  # Adjust the font size for better visibility
sns.heatmap(
    corr_matrix_debt,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    xticklabels=corr_columns,
    yticklabels=corr_columns,
)

# Set the title
plt.title("Correlation Heatmap for Debt-related Columns")

# Adjust plot settings
plt.tight_layout()

# Save the heatmap as an image file (e.g., for use in a presentation)
plt.savefig("debt_correlation_heatmap.png", dpi=300)

# Show the plot (optional)
plt.show()
```
